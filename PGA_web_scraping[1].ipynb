# Average driving distances on the PGA tour: 1980-2019

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import os
from matplotlib.offsetbox import TextArea, DrawingArea, OffsetImage, AnnotationBbox
import matplotlib.pyplot as plt
from matplotlib.pyplot import hist as hst
from matplotlib.pyplot import figure
```

# Get URL's for every tournament of every year

```python
urls=[]
Tournament=[]

starting_time=time.time()

for i in range(1980,2020):
    
    # Read page for year i
    
    url_1=f"https://www.pgatour.com/content/pgatour/stats/stat.101.y{i}.html"
    req=requests.get(url_1)
    html=req.content
    soup=BeautifulSoup(html)
    container=soup.select('div.statistics-details-select-wrap > div.with-chevron > select > option[value^="t"]')
    
    comps=[]
    
    for id in container:
        if str(id)[15:19]!='d="s':
            comps.append(str(id)[15:19])
        else:
            comps.append(str(id)[35:39])
    
    for k in comps:
        start_time=time.time()
        if k=='d="s':
            print("BAD URL")    
        else:
            url=f"https://www.pgatour.com/content/pgatour/stats/stat.101.y{i}.eon.{k}.html"
            #req2=requests.get(url)
            urls.append(url)
            Tournament.append(container[comps.index(k)].text)
            print(container[comps.index(k)].text,k)
            print(url,"Time:",start_time-time.time())

print("Final time:",starting_time-time.time())

save_dict={"URL":urls,"Tournament":Tournament}
save_urls=pd.DataFrame(save_dict)
save_urls.to_csv("C:\\Users\\victo\\00_Nod_coding_bootcamp\\Week 3\\Project\\urls_with_tournaments_TOURONLY.csv")
```

# Loop over URL's and get stats

```python
url_list=urls.copy()
url_list.reverse

rank=[]
rounds=[]
drives=[]
players=[]
total_dist=[]
total_drives=[]
year=[]
tournament=[]

starting_time=time.time()

for url_i in url_list:
    
    time.sleep(0.5)
    timer=time.time()
    req=requests.get(url_i)
    print("Loading URL")
    
    if str(req)=='<Response [200]>':
        
        print(url_i[56:60],"Bulding:",Tournament[url_list.index(url_i)])
        html=req.content
        soup=BeautifulSoup(html)
        rows=soup.select('[id^="playerStatsRow"]')
        
        for i in rows:
            #rank.append((i.select("td:nth-child(1)")[0].text.replace("\n","").replace(" ","")))
            rounds.append(float(i.select("td:nth-child(4)")[0].text))
            drives.append(float(i.select("td:nth-child(5)")[0].text.replace(",","")))
            players.append(i.select("a")[0].text)
            total_dist.append(float(i.select("td:nth-child(6)")[0].text.replace(",",".")))
            total_drives.append(float(i.select("td:nth-child(7)")[0].text))
            year.append(int(url_i[56:60]))
            tournament.append(Tournament[url_list.index(url_i)])
            
        
            
    else:
        continue

    print("Done, elapsed time:",time.time()-timer,"Total time:",(time.time()-starting_time)/60,"min")
```

# Compile all stats and save to a DataFrame

```python
stats={"Player":players,
        "Avg_Drive":drives,
        "Total_Dist":total_dist,
        "Total_Drives":total_drives,
        "Tournament":tournament,
        "Year":year} 

df = pd.DataFrame(stats)
df.to_csv("C:\\Users\\victo\\00_Nod_coding_bootcamp\\Week 3\\Project\\pga_tour_player_stats_TOURONLY_3.csv")
df
```

# Plotting data

```python
# Calculate: avg drive/year
plot_year=[]
avg_drive_year=[]
std_dv_drive=[]
avg_tot_dist=[]
avg_tot_drives=[]

for i in range(0,42):
    plot_year.append(1980+i)
    avg_drive_year.append(list(df_years)[i][1]["Avg_Drive"].mean())
    std_dv_drive.append(list(df_years)[i][1]["Avg_Drive"].std())
    avg_tot_dist.append(list(df_years)[i][1]["Total_Dist"].mean())
    avg_tot_drives.append(list(df_years)[i][1]["Total_Drives"].mean())
    
```

```python
figure(figsize=(15,7.5))
plt.grid(b=True)
plt.plot(plot_year,avg_drive_year)
plt.title("Average driving distance/year")
plt.xlabel("Year")
plt.ylabel("Distance [yds]")
plt.savefig('C:\\Users\\victo\\00_Nod_coding_bootcamp\\Week 3\\Project\\drive_all.png')

figure(figsize=(15,7.5))
plt.grid(b=True)
plt.plot(plot_year,std_dv_drive)
plt.title("Standard deviation of drives on tour per year")
plt.xlabel("Year")
plt.ylabel("Std. deviation [yds]")
plt.savefig('C:\\Users\\victo\\00_Nod_coding_bootcamp\\Week 3\\Project\\std_dev_all.png')

figure(figsize=(15,7.5))
plt.grid(b=True)
plt.plot(plot_year,avg_tot_dist)
plt.title("Average total distance/year")
plt.xlabel("Year")
plt.ylabel("Distance [1000-yds]")
plt.savefig('C:\\Users\\victo\\00_Nod_coding_bootcamp\\Week 3\\Project\\avg_tot_dist_all.png')

figure(figsize=(15,7.5))
plt.grid(b=True)
plt.plot(plot_year,avg_tot_drives)
plt.title("Average number of total drives/year")
plt.xlabel("Year")
plt.ylabel("No of drives")
plt.savefig('C:\\Users\\victo\\00_Nod_coding_bootcamp\\Week 3\\Project\\avg_tot_no_all.png')
```
